##Forward propogation:- z=$W^{T}*X$
=$\begin{bmatrix}
w_{0}\\ 
w_{1}\\ 
w_{2}\\
w_{3}\\
\end{bmatrix}*\begin{bmatrix}
1 & x_{1} & x_{2} & x_{3}
\end{bmatrix}$

From here we can observe that we will also get some bias whose weight is also going to update and we can take any value for bias.

Then, it will pass to activation function and activation fucntion is step function.
we will calculate  error=y-y_hat
Back Propogation:- We will use **perceptron weight update rule**

w=w_old+$\Delta w$

where $\Delta w = eta*X*error$

eta=learning rate

X=input

error =y-y_hat




##Lets understand why it's used for linear separataton only##

```python
activationfunc=lambda input,weight:np.where(np.dot(input,weight)>=0,1,0)

class Perceptron:
  def __init__(self,eta,epochs,activationfunc):   #here eta is learning rate which we can defined and epochs is how many time we want to repeat the process
    self.weight=np.random.randn(3)*1e-4          # Intialising the randome weight and multiplyint it to make them smaller
    self.eta=eta
    self.epochs=epochs
    self.activationfunc=activationfunc            

  def fit(self,x,y):                                                 #For training the model
    self.x=x
    self.y=y
    intialisation_with_bias=np.c_[self.x,np.ones((len(self.x),1))]       #concatenating the bias and input as we know that we can add some bias to translate it we can intialised this with any value
    for epoch in range(self.epochs):                           
      y_hat=activationfunc(intialisation_with_bias,self.weight)      #predicting the value with the given weight to the input
    
      error=self.y-y_hat                                             # calculating how close we are from actual value
      
      self.weight=self.weight+self.eta*np.dot(intialisation_with_bias.T,error)  #Backpropagation and changing the weight according to error we getting
  
  def predict(self,x,y):                                              #For predicting the output
    input_with_bias=np.c_[x,np.ones((len(x),1))]
    return self.activationfunc(input_with_bias,self.weight)
```

#**Back Propogation**

## lets understand back propogation from simple neural network which have 2 hidden layer with 1 unit each and sigmoid as activation function and loss function as $(y-yhat)^{2}$
## So after passing from first layer we get the result as

z1= $w_{1}^{[1]}*x_{1}+w_{2}^{[1]}*x_{2}+b$

a1=$\sigma(z1)=\frac{1}{1+e^{-z1}}$

z2=$a1*w^{[2]}+b$
________________________________________
a2=$\sigma(z2)=\frac{1}{1+e^{-z2}}$

z3=$a2*w^{[3]}+b$

_____________________________________

a3=$\sigma(z3)=\frac{1}{1+e^{-z3}}$

yhat=a3

Cost=$(y-yhat)^{2}$
______________________________________

## The differntiation of sigmoid function
$f(x)=\frac{1}{1+e^{-x}}$

$\frac{\mathrm{d} f(x)}{\mathrm{d} x}$=f(x)*(1-f(x))

Let's the get the rate of change in cost function wrt to $w^{[3]}$

$\frac{\mathrm{d}Cost }{\mathrm{d} w^{[3]}}=\frac{\partial cost }{\partial a3}*\frac{\partial a3 }{\partial z3}*\frac{\partial z3 }{\partial w^{[3]}}$

$\frac{\partial cost }{\partial a3}= 2*(y-a3)*(-1)  $

##Since relationship between a3 and z3 as sigmoid function so we will get it's derivative
$\frac{\partial a3 }{\partial z3}$=f(z3)*(1-f(z3))

## Since relationship between z3 and w represent as z3=$a2*w^{[3]}+b$
$\frac{\partial z3 }{\partial w^{[3]}}=a2$

Let's combine above equation

$\frac{\mathrm{d}Cost }{\mathrm{d} w^{[3]}}=2*(y-a3)*(-1)*f(z3)*(1-f(z3))*a2 $

## we can update th weight with this as 
w_new = w_old - learning_rate*$\frac{\mathrm{d}Cost }{\mathrm{d} w^{[3]}}$
________________________

##lets get the rate of change wrt $w^{[2]}$


$\frac{\mathrm{d}Cost }{\mathrm{d} w^{[2]}}=\frac{\partial cost }{\partial a3}*\frac{\partial a3 }{\partial z3}*\frac{\partial z3 }{\partial a2}*\frac{\partial a2 }{\partial z2}*\frac{\partial z2 }{\partial w^{[2]}}$

$\frac{\partial z3 }{\partial a2}=w^{[3]}$

$\frac{\partial a2 }{\partial z2}$ = f(z2)*(1-f(z2))

$\frac{\partial z2 }{\partial w^{[2]}}$ =a1

$\frac{\mathrm{d}Cost }{\mathrm{d} w^{[3]}} = 2 * (y-a3) * (-1) * f(z3) * (1-f(z3))* a2 * w^{[3]}$* f(z2) * (1-f(z2)) * a1

w_new = w_old - learning_rate*$\frac{\mathrm{d}Cost }{\mathrm{d} w^{[2]}}$
_______________________

$\frac{\mathrm{d}Cost }{\mathrm{d} w^{[1]}}=\frac{\partial cost }{\partial a3}*\frac{\partial a3 }{\partial z3}*\frac{\partial z3 }{\partial a2}*\frac{\partial a2 }{\partial z2}*\frac{\partial z2 }{\partial a1}*\frac{\partial a1 }{\partial z1}*\frac{\partial z1 }{\partial w^{1]}}$

$\frac{\partial z2 }{\partial a1}=w^{[2]}$

$\frac{\partial a1 }{\partial z1}$ = f(z1)*(1-f(z1))

$\frac{\partial z1 }{\partial w^{[1]}}$ =x

$\frac{\mathrm{d}Cost }{\mathrm{d} w^{[3]}} = 2 * (y-a3) * (-1) * f(z3) * (1-f(z3))* a2 * w^{[3]}$* f(z2) * (1-f(z2)) * $w^{[3]}$ *f(z1)*(1-f(z1)) * x

w_new = w_old - learning_rate*$\frac{\mathrm{d}Cost }{\mathrm{d} w^{[1]}}$

```python
x=np.linspace(-10,10,100)
def plot_graph(x,y,ALPHA=0.6, 
               label_x = r"$x \rightarrow$", label_y=r"$act(x) \rightarrow$", 
               title=None,
               LABEL=None):  #here we are taking input with alpha which is use to show how transparent and label for x and y where r represent take raw value
    # defining the figure size          
    plt.figure(figsize=(7,5))
    # defining the horizontal line
    plt.axhline(y=0, color="black", linestyle="--", lw=2)
    # defining the horizontal line
    plt.axvline(x=0, color="black", linestyle="--", lw=2)
    plt.xlabel(label_x)
    plt.ylabel(label_y)
    plt.title(title)
    
    if LABEL != None:
        plt.plot(x, y, alpha=ALPHA, label=LABEL);
        plt.legend(fontsize=14)

    else:
        plt.plot(x, y, alpha=ALPHA);


def derivative(f, z, eps=0.000001):
    return (f(z + eps) - f(z - eps))/(2 * eps)
```

#**Relu function**

$$ReLU(x)= max(x,0)$$

$$where\ ReLU(x) \in (0, x),\\
and\ x \in [-\infty, +\infty]$$


#Pros
## As we can see in relu function it's partly linear in nature so we will get the convergence faster as it will be easy for computer to calculate for linear function##
## There is derivative value for every postive number in relu function that  solves the problem of gradient saturation problem
#Cons
##At point 0 it's not continuous that means it's also not differentiatble at that point that leads to jumping gradient.
##Whenver we get negative value it will directly eliminate them as it have 0 differentiation it lead to dying relu problem
## Relu function is not zero centric so it will not give results for negative value.

```python
relu=lambda x: np.where(x>=0,x,0)
plot_graph(x,relu(x),title='Relu function',LABEL=r"$ReLU(x)= max(x,0)$")
plot_graph(x,derivative(relu,x),LABEL=r"$ReLU^\prime(x)$",title='Derivative')
```

#**Leaky Relu**

$$ 
leaky\_relu(x, \alpha) = \left\{\begin{matrix} 
x & x\geq 0 \\ 
\alpha x & x \lt 0 
\end{matrix}\right.
$$

$$where\ x \in [-\infty, +\infty]$$

## Leaky Relu solves the problem of dead gradient on when input are negative it give some gradient.

##It's same as relu but here we are using exponential term which is helping us to make it zero centric
## Dead Relu issue is solved here

#**Softmax function**

$$S(x_j)=\frac{e^{x_j}}{\sum_{j=1}^{K} e^{x_j}}, where\ j = 1,2, \cdots, K $$

## It gives us the probablity distribution of two or more than two category in the range between (0,1).
##It's mainly used in outer layer for multiclass classifcation as it provide the probability.

##Softmax function given by equation for k=2

$$p(y= 1|x) = \frac{1}{1+e^{-\theta^Tx}}$$

$$p(y= 0|x) = 1- p(y= 1|x)  =\frac{e^{-\theta^Tx}}{1+e^{-\theta^Tx}}$$

## Let say we have input as $\theta_0^Tx$  with assumpution of bias=0 and with 2 unit in output layer 
$$p(y= 1|x) = \frac{e^{\theta_1^Tx}}{e^{\theta_0^Tx} + e^{\theta_1^Tx}} = \frac{1}{1+e^{(\theta_0^T-\theta_1^T)x}} = \frac{1}{1+e^{-\beta x}}$$


$$p(y= 0|x) = \frac{e^{\theta_0^Tx}}{e^{\theta_0^Tx} + e^{\theta_1^Tx}} = \frac{e^{(\theta_0^T-\theta_1^T)x}}{1+e^{(\theta_0^T-\theta_1^T)x}} = \frac{e^{-\beta x}}{1+e^{-\beta x}}$$

##If we take $\theta_0^T-\theta_1^T =\beta$
## It will make the both equation look similar. Hence that proves that sigmoid function is equal to softmax at k=2
